```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    crime = pd.read_csv("crime_data.csv", index_col = 'Year')  #read in data from file. index is used as x axis for visualizations
    crime_list = crime.keys()  #get list of all names of columns
    
    #create 3 visualizations:
    #make a histogram of the crime data set: plot robbery crime as histogram
    crime.plot(kind = "hist", y = 'Robbery', color = "deeppink", edgecolor = "navy")
    
    #make a scatterplot:
    crime.plot(kind = "scatter", x = 'Murder', y = 'Forcible Rape', color = "mediumorchid")
    
    #make a line plot: put all the columns in it
    crime.plot(kind = "line", y = ['Violent', 'Property', 'Murder', 'Forcible Rape', 'Robbery', 'Aggravated Assault', 'Burglary', 'Larceny', 'Vehicle Theft'])
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    crime = pd.read_csv("crime_data.csv", index_col = 'Year')  #read in data from file. index is used as x axis for visualizations
    
    #make box plot for 1st one so that can see distribution of all 3 crime types:
    crime.plot(kind = "box", y = ['Burglary', 'Violent', 'Vehicle Theft'], color = "palevioletred")
    
    #make hexbin plot (since have 2 things plotting)for distributions of Vehicle Theft vs values of Robbery
    crime.plot(kind = "hexbin", x = "Vehicle Theft", y = "Robbery", gridsize = 20, sharex = False, colormap = "RdPu")
    
```
