```python
import numpy as np
from scipy.optimize import minimize as mini
from scipy.integrate import quad
from matplotlib import pyplot as plt
```

# Problem 1

Solve (16.2) for a(x).
Produce the plot shown in Figure 16.2.

```python
#define all variables they give us in prob:
c0 = 3/8
c1 = 5/4

xs = np.array([.1*j for j in range(11)])
u_prime = lambda x: 2*x + (1/2)

#get piecewise f function:
def f(x):
    if x <= (1/2):
        return -6*(x**2) + (3*x) - 1
    else:
        return -1
    
#solve u'(x_j) eq. under eq. 16.4 for alpha_j, can solve for alpha_j analitcally and then code it here!
alphas = [(c0 - quad(f, 0, x)[0]) / u_prime(x) for x in xs]  #only want 1st thing returned from quad method

#now plot the alphas:
plt.plot(xs, alphas, color = "dodgerblue")
plt.plot(xs, alphas, '*b')
plt.title("Solution a(x) for Problem 1", color = "deeppink")
plt.xlabel("x", color = "deeppink")
plt.ylabel("y", color = "deeppink")
plt.show()
```

# Problem 2

Find the density function a(x). Using several values of ε > 0.66049142, plot the corresponding
density a(x) for x in np.linspace(0,1,11) to demonstrate that the problem is ill-posed.

```python
#just need to change parameters from prob 1:

#define all variables they give us in prob:
c0 = 1
c1 = 2
eps_vals = [0.7, 0.8, 0.9]

xs = np.linspace(0, 1, 11)
f = lambda x: -1

#go through and do for all the eps values:
for eps in eps_vals:
    u_prime = lambda x: 1 + (1/eps)*np.cos((eps**(-2))*x)
    #solve u'(x_j) eq. under eq. 16.4 for alpha_j, can solve for alpha_j analitcally and then code it here!
    alphas = [(c0 - quad(f, 0, x)[0]) / u_prime(x) for x in xs]  #only want 1st thing returned from quad method
    
    #now plot the alphas:
    plt.plot(xs, alphas, label = "eps = " + str(eps))
    plt.title("Solution a(x) for Problem 1", color = "mediumvioletred")
    plt.xlabel("x", color = "mediumvioletred")
    plt.ylabel("y", color = "mediumvioletred")
plt.legend(loc = "best")
plt.show()
```
